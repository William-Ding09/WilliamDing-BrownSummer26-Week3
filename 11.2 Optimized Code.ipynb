```python
# Import necessary libraries
import yfinance as yf
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
```

```python
# This function downloads stock data and returns closing prices only.
def process_stock_data(ticker, start_date, end_date):
    
    """
    Downloads and processes stock data from Yahoo Finance.
    
    Parameters:
    ticker (str): Stock ticker symbol.
    start_date (str): Start date in 'YYYY-MM-DD' format.
    end_date (str): End date in 'YYYY-MM-DD' format.
    
    Returns:
    pandas.Series: Series of closing prices.
    """
    
    data = yf.download(ticker, start=start_date, end=end_date)
    closing_prices = data['Close']
    return closing_prices
```

```python
# Function to fit and predict using ARIMA
def arima_model(closing_prices):
    
    """
    Fits an ARIMA model to the closing prices and makes predictions.
    
    Parameters:
    closing_prices (pandas.Series): Series of closing prices.
    
    Returns:
    pandas.Series: Predicted values.
    """
    
    model = ARIMA(closing_prices, order=(5,1,0))
    model_fit = model.fit()
    predictions = model_fit.predict(typ='levels')
    return predictions
```

```python
# Function to plot the data
def plot_data(actual, predicted, title):
    
    """
    Plots actual vs predicted stock prices.
    
    Parameters:
    actual (pandas.Series): Actual closing prices.
    predicted (pandas.Series): Predicted closing prices.
    title (str): Title of the plot.
    """
    
    plt.figure(figsize=(10,4))
    plt.plot(actual, label='Original')
    plt.plot(predicted, color='red', label='Predicted')
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('Price ($)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def compute_mape(actual, predicted):
    
    """
    Computes the Mean Absolute Percentage Error (MAPE).
    
    Parameters:
    actual (pandas.Series): Actual values.
    predicted (pandas.Series): Predicted values.
    
    Returns:
    float: MAPE value.
    """
    
    mape = mean_absolute_error(actual, predicted)/len(actual) * 100
    return round(mape, 2)
```

```python
# Start and end dates
start_date = '2019-01-01'
end_date = '2023-07-21'

# Define tickers (just change these!)
ticker1 = 'MSFT'
ticker2 = 'AMZN'

# Process stock data
closing_prices_1 = process_stock_data(ticker1, start_date, end_date)
closing_prices_2 = process_stock_data(ticker2, start_date, end_date)

# Fit ARIMA and make predictions
predictions_1 = arima_model(closing_prices_1)
predictions_2 = arima_model(closing_prices_2)

# Compute MAPE
mape_1 = compute_mape(closing_prices_1, predictions_1)
mape_2 = compute_mape(closing_prices_2, predictions_2)

# Print MAPE
print(f"MAPE for {ticker1}: {mape_1}%")
print(f"MAPE for {ticker2}: {mape_2}%")

# Plot the data
plot_data(closing_prices_1, predictions_1, f'{ticker1} Stock Price: Original vs. Predicted')
plot_data(closing_prices_2, predictions_2, f'{ticker2} Stock Price: Original vs. Predicted')
```

```python
# List of tickers you want to analyze
tickers = ['AAPL', 'TSLA', 'MSFT', 'AMZN']  # Just add or remove tickers here

# Start and end dates
start_date = '2019-01-01'
end_date = '2023-07-21'

# Loop through each ticker
for ticker in tickers:
    print(f"\nProcessing {ticker}...")

    # Step 1: Get closing prices
    closing_prices = process_stock_data(ticker, start_date, end_date)

    # Step 2: Fit ARIMA model and get predictions
    predictions = arima_model(closing_prices)

    # Step 3: Compute MAPE
    mape = compute_mape(closing_prices, predictions)
    print(f"MAPE for {ticker}: {mape}%")

    # Step 4: Plot actual vs. predicted
    plot_data(closing_prices, predictions, f'{ticker} Stock Price: Original vs. Predicted')
```
